```python
import pandas as pd
import matplotlib.patches as mpatches


df = pd.DataFrame([
    [  8,  448.0,  70.94,       'RTX3070', 'NVIDIA'],
    [ 10,  760.3, 106.40,       'RTX3080', 'NVIDIA'],
    [ 12,  912.4, 106.72,       'RTX3080', 'NVIDIA'],
    [ 12,  504.2,  82.21,       'RTX4070', 'NVIDIA'],
    [ 16,  716.8, 106.22,       'RTX4080', 'NVIDIA'],
    [ 20,  360.0,  58.59,    'RTX4000Ada', 'NVIDIA'],
    [ 24,  936.2, 111.74,       'RTX3090', 'NVIDIA'],
    [ 24, 1010.0, 127.74,       'RTX4090', 'NVIDIA'],
    [ 32,  576.0,  89.87,    'RTX5000Ada', 'NVIDIA'],
    [ 48,  768.0, 102.22,      'RTXA6000', 'NVIDIA'],
    [ 48,  960.0, 130.99,    'RTX6000Ada', 'NVIDIA'],
    [ 48,  695.8,  88.95,           'A40', 'NVIDIA'],
    [ 48,  864.0, 113.60,          'L40S', 'NVIDIA'],
    [ 80, 1940.0, 138.31,      'A100PCIe', 'NVIDIA'],
    [ 80, 2040.0, 133.38,       'A100SXM', 'NVIDIA'],
    [ 80, 2040.0, 144.49,      'H100PCIe', 'NVIDIA'],
    [  8,   68.3,   9.72,       'M17Core',  'APPLE'],
    [ 64,  409.6,  34.49,   'M1Max32Core',  'APPLE'],
    [192,  819.2,  76.28, 'M2Ultra76Core',  'APPLE'],
    [ 64,  409.6,  50.74,   'M3Max40Core',  'APPLE'],
], columns=['memory', 'bandwidth', 'generation speed', 'type', 'manufacturer'])

g = df.plot.scatter(
    x='bandwidth',
    xlabel='bandwidth (GB/s)',
    y='generation speed',
    ylabel='generation speed (tokens/s)',
    c=df['manufacturer'].apply(lambda x: { 'APPLE': 'red', 'NVIDIA': 'blue' }[x]),
)

for k, v in df.iterrows():
    g.annotate(
        v[3],
        xy=(v[1], v[2]),
        size=7.5,
        xytext=(0,-10),
        textcoords='offset points'
    )

g.legend(
    handles=[mpatches.Patch(color='red', label='Apple'), mpatches.Patch(color='blue', label='Nvidia')],
    loc='upper left'
)
```
